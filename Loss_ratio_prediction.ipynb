```python
def createInstances(df,df_imp):
    col = []
    val = []
    for column in df.columns:
        total = df[column].count()
        if(column=='Vehicle_Miles_To_Work'):
            col.append(column)
            val.append(df[column].mean())
        if(column=='Annual_Premium'):
            col.append(column)
            val.append(df[column].sum())
        if(column=='Loss_Amount'):
            col.append(column)
            val.append(df[column].sum())
        if(column!='Index' and column!='Vehicle_Miles_To_Work' and column!='Annual_Premium'and column!='Loss_Amount'):
            contents = df[column].unique()
            for c in contents:
                col.append("Percentage_Of_" + str(c) + "_" + column)
                val.append((df[df[column] == c].count()['Index']/total)*100)
        
    
    df_imp = pd.concat([df_imp,pd.DataFrame(columns=col)])
    df_imp = df_imp.append(pd.Series(val, index=col), ignore_index=True)
    return df_imp
```

```python
df_toTrain = pd.DataFrame()

path = 'training_portfolios'
all_files = glob.glob(os.path.join(path, "*.csv"))

for file in all_files:    
    
    df1000 = pd.read_csv(file)
    df1000 = df1000.reset_index(drop=True)
    df1000 = df1000.drop(['PolicyNo','Unnamed: 0'], axis=1)
    df1000['Index'] = df1000.index
    df_toTrain = createInstances(df1000,df_toTrain)
    
# seed(1)

# filename = "training_portfolio_main_" + str(randint(0, 1000)) + ".csv"
# df_toTrain.to_csv(filename)
```
